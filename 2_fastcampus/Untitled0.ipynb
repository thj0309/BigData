```
# 사전 자료형 관련 함수
data = dict()
data['apple'] = '사과'
data['banana'] = '바나나'
data['coconut'] = '코코넛'
```

```
# 사전 자료형 관련 함수 2
key_list = data.keys()
print(key_list)
```

```
#사전 자료형 관련 함수 3
# 값 데이터만 남은 리스트
value_list = data.values()
print(value_list)
```

```
for key in key_list:
  print(key + " : " + data[key])
```

```
# 집합 자료형 소개
# 초기화 방법 1
data = set([1,1,2,2,3,4,5,6])
print(data)
```

```
# 집합 자료형 소개
# 초기화 방법 2
data = {1,1,2,2,3,4,5,6}
print(data)
```
