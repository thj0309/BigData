```
# 하나의 torch 만들기.
V_data = [1., 2., 3.]
V = torch.Tensor(V_data)
print(V)
```

```
# torch.ones : tensor를 size에 맞게 1로 채운다
a = torch.ones(3)
print(a)
```

```
# torch.zeros : tensor를 size에 맞게 0로 채운다
b = torch.zeros(3)
print(b)
```

```
# 초기화되면서 현재 메모리에 존재하는 값
x = torch.empty(5,3)
print(x)
```

```
# rand값 넣기
rand = torch.rand(5,3)
print(rand)
```

```
# torch.rand : 랜덤 숫자로 tensor를 채운다
rand_points = torch.rand(4, dtype=torch.double)
print(rand_points)

rand_points_2 = torch.rand(2,3,dtype=torch.float)
print(rand_points_2)
```

```
# 행렬 사이즈
print(rand_points_2.size())
```

```
# 2. Indexing Tensors
temp = torch.Tensor([0,1,2,3,4,5,6,7])

print(temp[:]) # tensor element 전체
print(temp[1:4]) # 1부터 4까지
print(temp[1:]) # 1부터 끝까지
print(temp[:4]) # 0부터 4까지
print(temp[-1]) # 맨뒤 한 개 값
print(temp[1:6:2]) # index 1과 6사이의 원소 출력 (2단계씩 증가)
```

```
temp2 = torch.Tensor([[1,2,3],[4,5,6]])

print(temp2)
```

```
# 3. tensor element types
## 
float_points = torch.ones(5,2,dtype=torch.float32)
int_points = torch.zeros(3,dtype=torch.int64)

print(float_points)
print(int_points)
```

```
# default - float
default_points = torch.Tensor([2,2])
print(default_points)
```

```
# find out dtype
print(float_points.dtype)
print(int_points.dtype)
print(default_points.dtype)
```

```
# other fomrs of specifying dtype
to_int_1 = float_points.int()
print(to_int_1)

# 'to' check
to_int_2 = float_points.to(torch.int64)
print(to_int_2)

to_float_1 = int_points.float()
print(to_float_1.dtype)

to_float_2 = int_points.to(dtype=torch.float16)
print(to_float_2.dtype)
```

```
# 4. the tensor api
# 5. tensor shape and reshape
```

```
# shape : 텐서 사이즈 보여주기
a = torch.Tensor([[1,2,3],[4,5,6]])
print(a)
print(a.shape)
```

```
# size()
print(a.size())
```

```
# reshape : 주어진 shape로 original tensor에서 데이터를 가져와서 reshape한다
a_reshape=a.reshape(3,2)
print(a_reshape)
print(a_reshape.shape)

a_reshape2=a.reshape(-1,2)
print(a_reshape2)
print(a_reshape2.shape)

```

```
a_no_reshape = a.reshape(4,3)
# element 갯수가 맞지 않음  
```

```
# view : returns a new tensor with the same data
x = torch.randn(4,4)
print(x.size())

y = x.view(16)
print(y.size())


z = x.view(-1, 8)
print(z.size())
# 행렬 변환 하는 기능
```

```
print(y)
```

```
print(z)
```

```
# 6. transpose
## view와는 다른 개념이다
x  = torch.randn(1)
print(x)
print(torch.t(x))
```

```
y = torch.randn(3)
print(y)
print(y.t())
```

```
z = torch.randn(2,3)
print(z)
print(torch.t(z))
```

```
## torch.transpose
w = torch.randn(2, 3)
print(w)
print(torch.transpose(w,0,1))
```

```
# more complex case
real_example = torch.randn(64, 128, 768)
print(real_example.shape)
transpose = torch.transpose(real_example, 1,2)
print(transpose.shape)
```

```
# 7. operations example
x = torch.ones(1,2)
print(x)
y = torch.ones(5,3)
print(y)
```

```
print(x+y)
# x와 y는 사이즈가 맞지 않아 연산이 안된다
```

```
y = torch.ones(5,2)
print(x+y)
```

```
# 더하기 연산
print(x+y)
print(torch.add(x,y))

## 빈 공간 만들어서 연산한 후 집어넣기
result = torch.empty(5, 2)
torch.add(x, y, out=result)

print(y.add_(x))
```
