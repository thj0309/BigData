리스트, 행렬 선언

```
a = [1,2, 3]
b = ['a', 'b', 'c']
c = [['a'], ['a','b'], ['a','c']]

print (a)
print (b)
print (c)
```

```
a = np.array([1,2, 3])
b = np.array(['a', 'b', 'c'])
c = np.array([['a'], ['a','b'], ['a','c']])

print (a)
print (b)
print (c)
```

```
# appendfh item 추가하기
#c.append('new item')
c.append(['new item'])

print (c)
```

```
# 배열 또는 행렬에 1로 채우기
print(np.ones((1,5)))
print(np.ones((2,3)))
```

```
# 배열 또는 행렬에 0로 채우기
print(np.zeros((1,5)))
print(np.zeros((2,3)))
```

```
# 배열 또는 행렬 안에 순차적으로 증가하는 리스트 만들기
print(np.arange(10))
print(np.arange(3, 7, dtype=np.float))
print(np.arange(3, 7, 2))
```

```
# 행렬 선언하기
mat1 = np.array([[1,2,3],[4,5,6]])

# random으로 matrix 만들기
mat2 = np.random.randint(low=1, high=10, size=(3,2))
mat3 = np.random.rand(3,2)

print(mat1)
print(mat2)
print(mat3)
```

```
# 리스트 인덱싱(indexing) & 슬라이싱(slicing)
#indexing
a = [1,3,5,7,9,11]
print(a)

print(a[2], a[5], a[-1])
```

```
# slicing
b= [2,4,6,8,10]
print(b[2:])
print(b[:2])
print(b[:])
```

```
# numpy의 reshape(pytorch의 view와 비교)
# (row, col)
matt1 = np.random.rand(6,3)
print(matt1)
```

```
# -1: all
print(matt1.reshape(1,-1).shape)
print(matt1.reshape(1,-1))
print("---------------------------------")
print(matt1.reshape(-1,1).shape)
print(matt1.reshape(-1,1))
```

```
print(matt1.reshape(2,9).shape)
print(matt1.reshape(2,9))
```

```
print(matt1.reshape(2,5).shape)
print(matt1.reshape(2,5))
# matt1는 3,6으로 구성된 행렬이므로 에러가 발생함
```

```
# tensor 형태로 나타내기
print(matt1.reshape(3,2,3).shape)
print(matt1.reshape(3,2,3))
```

```
# matrix or tensor 형태에서도 slicing 가능하다.
matt2 = np.arange(24).reshape(-1,4)
print(matt2)
```

```
print(matt2[:1,:3])
print(matt2[3,0:2])

matt2[:1,:3]
```

```
matt2[3,0:2]
```

```
# 사칙연산
x = np.array([[1,3,5], [7,9,11], [13,15,17]])
y = np.array([[2,4,6], [8,10,12], [14,16, 18]])

print(x)
print(y)
```

```
# add
x + y
```

```
x-y
```

```
x*y
```

```

x/y
```

```
# 제곱
print(x **2 )
print(np.power(x, 2))
```

```
# 내적구하기
np.dot(x,y)
```

```
# 제곱근
np.sqrt(x)
```

```
# more on matrix operation
z1 = np.array([[2,2,2]])
z2 = np.array([[2,2]])

z1 * z2
```

```
# transpose
print(z1.T)
```

```
z3 = np.array([1,2,3])

print(np.dot(z1, z3))
```
